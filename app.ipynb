```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# csv dosyasını dataframe olarak oku
df = pd.read_csv('diabetes.csv')
df.columns= df.columns.str.strip().str.lower()
```

```python
# 5 diyabet hastasını getir
df[df['outcome']==1].head()
```

```python
# Diabet hastası olanların yaşa göre dağılımı ve grafiği
diabetes = df[df['outcome']==1]
grouped_age_data = diabetes.groupby('age')['age'].count().reset_index(name='count')
grouped_age_data.plot(x='age', y='count', kind='line')
plt.show()
```
